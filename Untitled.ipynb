```python
import torch
from torch.autograd import Variable

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('movie_data.csv')
```

```python
df
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()

df['Movie'] = le.fit_transform(df['Movie'])
df['Month'] = le.fit_transform(df['Month'])
```

```python
df.head()
```

```python
#the input data can be represented using 2 matrics 
#inputs and targets
```

```python
#Inputs(Movie , Month , Month , Day , Year)
inputs = df[['Movie', 'Month', 'Day', 'Year', 'Budget($M)']]
```

```python
targets = df[['Domestic Gross($M)',
       'Worldwide Gross($M)']]
```

```python
inputs = np.asarray(inputs)
```

```python
targets = np.asarray(targets)
```

```python
#Before building a model , we need to convert inputs and targets as pytorch tensors
```

```python
inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
```

```python
print(inputs)
print(targets)
```

```python
#weights and biases
w = torch.randn(2,5,requires_grad=True)
b = torch.randn(2,requires_grad =True)
```

```python
#Define the model
def model(x):
    x=x.float()
    return x @ w.t() + b
```

```python
preds = model(inputs)
print(preds)
```

```python
print(targets)
```

```python
#MSE loss
def mse(t1, t2):
    t1=t1.double()
    t2 = t2.double()
    diff = t1 - t2
    return torch.sum(diff*diff)/diff.numel()
```

```python
#computing loss
loss = mse(preds , targets)
print(loss)
```

```python
#computing gradient descent
loss.backward()
```

```python
print(w)
print(w.grad)
```

```python
w.grad.zero_()
b.grad.zero_()
print(w.grad)
print(b.grad)
```

```python
#generate predictions
preds = model(inputs)
print(preds)
```

```python
#calculate the loss
loss = mse(preds, targets)
print(loss)
```

```python
loss.backward()
```

```python
with torch.no_grad():
    w -= w.grad * 1e-5
    b -= b.grad * 1e-5
    w.grad.zero_()
    b.grad.zero_()
```

```python
preds = model(inputs)
loss = mse(preds , targets)
print(loss)
```

```python
#training for 10 epochs
for i in range(10):
    preds = model(inputs)
    loss = mse(preds,targets)
    loss.backward()
    with torch.no_grad():
        w -= w.grad * 1e-5
        b -= b.grad * 1e-5
        w.grad.zero_()
        b.grad.zero_()
    
```

```python
preds = model(inputs)
loss = mse(preds , targets)
print(loss)
```
